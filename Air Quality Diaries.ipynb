<b> Data Visualisation

```python
# correlation matrix for numerical columns using heatmap
# most of heatmap is yellow/ pale yellow (0,-1)  this indicates no linear relation between variables & negative correlation


import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df.select_dtypes(include = ['float64']).columns.tolist()
heatmap_matrix = df[num_cols].corr()

sns.heatmap(heatmap_matrix, annot=True, cmap="YlGnBu", linewidths=0.5, fmt= ".1f")
plt.title("Correlation Matrix", fontweight = "bold", pad = 20 )
plt.show()
```

```python
# The AQI scale typically ranges from 0 to 500
# 0 to 50: Good
# 51 to 100: Moderate
# 101 to 150: Unhealthy for sensitive groups. 
# 151 to 200: Unhealthy. 
# 201 to 300: Very unhealthy. 
# 301 and above: Hazardous
# AQI for Bangalore is around 98 - moderate range

sns.histplot(data= df, y ="AQI", color = "purple")
plt.title("Distribution of Quality Air Index", fontweight = "bold", pad = 20)
plt.show()
```

```python

sns.histplot(data= df, y ="NO2(ug/m3)", color = "purple")
plt.title("Distribution of NO2(ug/m3)", fontweight = "bold", pad = 20)
plt.show()
```

```python
for column in df.columns:
    sns.scatterplot(x=df[column],y=df["AQI"], color= "purple")
    
    plt.xticks(rotation=45)
    plt.show()
```
